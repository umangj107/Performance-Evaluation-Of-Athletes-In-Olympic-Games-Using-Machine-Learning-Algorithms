```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```
data= pd.read_csv('/content/drive/My Drive/DataMiningProjecct/Preprocessed-datasets/olympics_final_final.csv')
```

```
data= data[data['Medal_Won']!=0]
data1=pd.DataFrame(data[['Age','Weight','Height','Medal_Won']])
#print(data.head())
#print(data['Age'].describe())
```

```
kmeans = KMeans(n_clusters=3).fit(data1)
centroid = kmeans.cluster_centers_
print(centroid)
```

```
plt.scatter(data1['Weight'],data1['Height'],c=kmeans.labels_.astype(float),s=50,alpha=0.5)
plt.scatter(centroid[:,1],centroid[:,2], c='red',s=50)
plt.show()
```
